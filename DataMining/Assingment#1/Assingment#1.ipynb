```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#0. csv파일 읽기
df2=pd.read_csv('C:/Users/ekdms/Documents/My-Python/Assingment#1_result.csv', 
                encoding='ANSI', index_col=0,engine='python')
df2.columns=['first','second','third','fourth']

#1.데이터 준비
x=list(df2.columns)
year=2015

#2. x축과 y축 데이터를 지정하여 라인플롯 생성
for i in range(0,6):
    y=list(df2.loc[year+i])
    plt.plot(x,y,label=year+i)
    
#3. 차트 제목 설정
plt.title('2015~2020 Quarterly sales')

#4. x축 레이블 설정
plt.xlabel('Quaters')

#5. y축 레이블 설정
plt.ylabel('sales')

#6. 항목 레이블 지정
plt.legend(list(df2.index))

#7.라인플롯 표시
plt.show()
```
