```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import datetime
import ssl

def Barun_store(result):
    for page in range(1,16):
        Barun_url = 'https://barunchicken.com/store/?pageid=%d&mod=list'%page
        #바른치킨의 지점의 정보를 알려주는 홈페이지 url입니다.
        #페이지 번호는 'page'에 저장이 되고 1~15페이지의 내용을 크롤링합니다.
        print(Barun_url)
        context = ssl._create_unverified_context()
        #SSLCertVerificationError를 해결하기 위한 부분입니다.
        html = urllib.request.urlopen(Barun_url, context=context)
        soupBarun = BeautifulSoup(html, 'html.parser')
        #soupBarun에 해당 해당 url의 내용을 저장합니다.
        ul_class = soupBarun.find('ul',class_ = "kboard-list")
        #soupBarun에서 'ul'중 클래스 명이 "kboard-list"인 부분을 ul_class에 저장합니다.
        for store in ul_class.find_all('li'):
            #ul_class에서 'li'인 부분을 store에 저장합니다.
            if len(store) <= 1: #길이가 1 이하인 경우 크롤링을 종료합니다.
                break

            store_name = store.find('div', class_="kboard-worldmap-franchise-branch")
            #<div class="kboard-worldmap-franchise-branch">지점명 : 목포대성점</div> 형태로 저장
            store_name = store_name.string #split을 사용하기 위해 string으로 바꿔줍니다.
            store_name=store_name.split('지점명 : ') #'지점명 : '을 기준으로 나눠줍니더. 
            store_name = store_name[1] # 지점의 이름만을 'store_name'에 저장합니다.
            
            store_local = store.find('div', class_="kboard-worldmap-franchise-area")
            #<div class="kboard-worldmap-franchise-area">지역 : 전라남도</div>형태로 저장
            store_local = store_local.string
            store_local=store_local.split('지역 : ')
            store_local = store_local[1]
            
            store_address = store.find('div', class_="kboard-worldmap-franchise-address")
            #<div class="kboard-worldmap-franchise-address">주소 : 전남 목포시 마파지로27번길 10</div>형태로 저장
            store_address = store_address.string
            store_address = store_address.split('주소 : ')
            store_address = store_address[1]
            
            store_phone = store.find('div', class_="kboard-worldmap-franchise-tel")
            #<div class="kboard-worldmap-franchise-tel">연락처 : <a href="tel:061-272-5850" title="061-272-5850">061-272-5850</a></div>형태로 저장
            store_phone = store_phone.find('a') #store_phone중에서 'a'에 해당되는 부분을 저장합니다.
            store_phone = store_phone.string #store_phone을 string으로 바꿔줍니다.            

            result.append([store_name]+[store_local]+[store_address]+[store_phone])
            # 지점명(store_name), 지역(store_local), 주소(store_address), 전화번호(store_phone)를 result에 저장합니다.
    return

def main():
    result = []
    print('Barun store crawling >>>>>>>>>>>>>>>>>>>>>>>>>>')
    Barun_store(result) 
    Barun_tbl = pd.DataFrame(result, columns = ('store', 'local', 'address', 'phone'))
    #지점의 정보를 담을 데이터프레임을 만듭니다.
    Barun_tbl.to_csv("C:/Users/ekdms/Codes/My_Python/Task02/Barun.csv", encoding = "cp949", mode = "w", index = True)
    #Barun_tbl을 csv파일로 저장합니다.
    del result[:]
    
if __name__ == '__main__':
    main()
```
